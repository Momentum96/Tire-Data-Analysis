```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

def bandpassFiltering(
    df: pd.DataFrame,     # Input DataFrame
    col: list,            # List of column names to apply the filter on
    lowcut: float,        # Low frequency cutoff
    highcut: float,       # High frequency cutoff
    fs: int,              # Sampling frequency
    order=5,              # Order of the Butterworth filter (default is 5)
    inplace: bool = False # Modify the DataFrame in-place or create new filtered columns (default is False)
):
    nyq = 0.5 * fs  # Calculate Nyquist frequency
    low = lowcut / nyq
    high = highcut / nyq

    for i in col:
        # Select appropriate filter type based on low and high cutoff frequencies
        if low == 0:
            b, a = butter(order, high, btype="low")
        elif high == 0:
            b, a = butter(order, low, btype="high")
        else:
            # Compute filter coefficients for a bandpass filter
            b, a = butter(order, [low, high], btype="band")
        
        # Apply the filter to each specified column in the DataFrame
        if not inplace:
            # Create new columns with filtered data
            df[i + "_filtered"] = filtfilt(b, a, df[i].to_numpy())
        else:
            # Modify original columns with filtered data
            df[i] = filtfilt(b, a, df[i].to_numpy())

    return df
```

```python
"""
    생각의 흐름
    (1) Spectrogram 이미지를 봤을 때, 포트홀을 밟은 순간에는 대체적으로 Contact Patch에 해당하는 저주파 영역을 벗어나는 고주파 영역에서의 신호가 두드러지게 관찰됨.
    (2) 시간 도메인 신호를 관찰할 경우 포트홀을 밟은 순간에 대한 판단이 모호하나, 주파수 도메인, 특히 Spectrogram 이미지로 관찰했을 때 포트홀을 밟은 순간에 대해 판단이 보다 직관적임.
    (3) Road Classification CNN Model은 Spectrogram 이미지를 입력받아 도로를 분류하는 형태이며, GUI 프로그램 상에서 실시간으로 동작할 수 있게 구현한 자료가 있음.
    
    (4) 초기 포트홀 감지에 대한 아이디어는 왼쪽과 오른쪽 양 바퀴 간 주파수 도메인 신호의 차이를 토대로 Threshold를 설정하여 특정 크기 이상의 차이가 발생했을 시
        더 큰 크기의 값을 갖는 쪽의 바퀴가 포트홀을 밟은 것으로 판단하는 형태를 예상.
    (5) 위 아이디어를 실현하기 위해서는 포트홀을 밟아서 고주파 신호가 발생한 그 시간을 최대한 정확히 구분하고, 그 영역에 포함되는 sample 크기만큼의 양쪽 바퀴 센서 신호를
        주파수 도메인으로 분석하여 차이를 비교하는 것이 이상적
    
    (6) 시계열 신호에서 해당 아이디어를 구현하려면 대략적인 생각으로는 Contact Patch에 해당하는 저주파보다 큰 범위의 신호들만 남기기 위한 high pass filtering과,
        어느 고주파 신호 크기를 포트홀을 밟았다고 볼 것인지에 해당하는 가속도 크기에 대한 Threshold가 필요
    (7) 가속도의 Range는 속도에 대한 dependency가 있음. 이는 즉 포트홀을 밟았다고 가정하더라도 속도에 따라 그 Threshold 값이 달라질 수 있음을 의미함.
    (8) Contact Patch Extraction 알고리즘을 주파수 도메인으로 변환하면서 가져갈 수 있는 이점 중 내가 생각하는 가장 큰 장점은 가속도 크기를 사용한 Threshold가 아닌
        Contact Patch에 의해 발생하는 저주파 볼록 신호를 0~500Hz의 주파수 영역 전체의 크기로 나누어줌으로써 현재 속도에 따른 가속도 Range 변화와 관계없이 현재 주파수 크기의 비율을 통해
        Contact Patch를 판단할 수 있다는 장점.
        (실제 이를 통해 30km/h 이하의 속도인 10km/h 속도에서도 Contact Patch 추출에 대한 가능성이 보여짐)
    (9) Spectrogram 이미지 또한 수집된 가속도 값의 Range에 영향을 받지 않으며, 상대적인 크기를 통해 그 주파수 대역에서의 크기가 색상으로 표현되는 것
    
    (10) 따라서, 실제 사람이 보았을 때 직관적으로 확인이 되는 Spectrogram 이미지를 통해 포트홀을 밟은 시점을 파악할 수 있으면 좋겠다는 생각이 들게됨.
        (양하늘 박사님과 포트홀 주행시험 데이터 분석 내용에 대해 이야기해보면서 실시간 알고리즘으로 적용할 때 까지 고려했을 때 해당 방법이 유리할 것으로 판단됨)
    (11) 포트홀에 의해 관찰되는 고주파 영역 신호도 있지만, 이를 이상적이지 않은 일반 도로 상황에서 울퉁불퉁함에 의해 발생하는 노이즈와 어떻게 구분해야 할지 감이 잘 잡히지 않음
    (12) 이를 해결하기 위한 방법을 생각해보았고, 아래는 그 아이디어를 구현해본 코드.


    알고리즘 개요
    1. Spectrogram 이미지 생성 (하단 코드부터는 OpenCV를 사용하여 Spectrogram 이미지 생성, OpenCV는 이미지를 numpy array로 다루기에 속도가 빠르며 이후 이미지를 사용한 연산에 있어서도 속도가 보장됨)
        (OpenCV에서 표현되는 이미지는 좌측 상단을 기준으로 우측으로 가면서 x축이 증가하고, 하단으로 가면서 y축이 증가하는 형태임)
        (Spectrogram의 주파수 범위에 해당하는 y축은 실제 표기 시 아래를 0으로 표기하므로, 이미지 또한 표시 형식을 이에 맞추도록 뒤집음)
    
    2. 구간 나누기
        Road Classification CNN Model의 경우, 500x500 크기의 0.5초에 해당하는 긴 시간에 대한 Spectrogram 이미지를 넣어주었는데 이에 따라 사용된 모든 신호가 학습에 사용됨
        포트홀 검출의 경우 그 시간을 0.5초로 하던 전체 신호로 하던, 가장 중요한 부분은 포트홀을 밟은 시점인 아주 짧은 시간간격에 해당하는 신호가 중요함
        
        단순히 Z축 가속도의 Spectrogram 이미지를 보았을 때, 이 이미지를 특정 간격으로 항상 자를 수 있는 방법에 대한 직관은 역시 저주파 Contact Patch 신호만한게 없음
        하지만 실시간으로 이 Contact Patch 위치를 찾자고 기존에 만들어진 무선전력으로 센서가 위치하는 각도 구하기 -> Contact Patch가 포함되는 예상 각도 범위 -> 주파수 분석 -> Contact Patch 추출하는 과정을 거치기에는
        Spectrogram 이미지 상에서 너무 확연하게 보이기에 굳이 그렇게까지 해야하나? 연산량이 아깝다는 생각이 듬
        (물론 초기에 Contact Patch Extraction 알고리즘을 만들 때는 연산량에 대해 고려가 되지 않았고, 충분히 발전시킬수 있다고 생각은 하지만 여기에서는 최대한 Spectrogram 이미지 상의 정보만을 사용하는 것이
        연산량을 최소화할 수 있는 방법이라고 생각이 들었습니다.)
        
        그래서 Spectrogram 이미지 상에서 하얗게 보이는 저 구간의 중심점의 X축 좌표(시간)만 추출하는 알고리즘이 있으면, 속도가 어찌되었든 저 저주파 신호는 발생을 할 것이니 이를 토대로
        포트홀을 밟았을 때 발생하는 고주파 신호를 탐색하기 위한 세트로 나누는 기준이 될 수 있지 않을까? 라는 생각을 하게 됨
        
    2-1. Low Pass Filtering
        위 2에서 결론대로 구간을 나눈다고 했을 때, 항상 반복적으로 발생하는 Contact Patch에 해당하는 저주파 신호의 위치를 이미지 상에서 추출할 수 있는 방법을 생각
        기존 Contact Patch Extraction 발표자료 중 Low Frequency Range라는 파라미터를 정의하기 위해 서로 다른 속도에서의 주파수 신호 분포 양상을 확인하였음. (발표자료 25페이지 참고)
        해당 알고리즘에서는 볼록한 신호의 끝나는 지점이 중요했으나, Spectrogram에서는 가장 큰 크기를 갖는 주파수 영역이 가장 밝게 나타날 것임
        전륜 바퀴 기준 10~110km/h 속도에서 수집한 데이터를 분석하였을 때, 그 피크 신호는 100Hz 안에서 존재하였음
        따라서, Spectrogram 이미지 중 100Hz 이상 범위의 값은 사실상 Contact Patch에 해당하는 신호를 구분하는데 있어서 필요가 없으므로, 100Hz 이하 Low Pass filter를 적용한 신호를 사용하여
        고주파 노이즈가 존재하지 않는 Spectrogram을 생성하는 것이 더 유리할 것이라 판단
        
    2-2. 픽셀 값 이진화
        Low Pass Filtering이 적용된 Spectrogram 이미지에서는 대부분이 검정색, Contact Patch에 해당ㅎ는 부분만 흰색으로 나타날 것임
        구간을 나누기 위한 핵심은 픽셀의 값이 아닌, Contact Patch인 구간과 아닌 구간을 구분하는 알고리즘이 필요
        이를 위해 0~255 사이의 값을 갖는 픽셀을 특정 threshold값을 기준으로 그 이상이면 255, 그 미만이면 0으로 이진화하는 방식을 적용
        
    2-3. Y축 병합
        Spectrogram 이미지는 X축은 시간 정보, Y축은 주파수 간격정보, 픽셀 값은 해당 시간의 해당 주파수 영역의 크기를 나타냄
        Low Pass Filtering과 픽셀 값 이진화를 적용하였을 때, 그 주파수에서의 값이 중요한 것이 아니고 Contact Patch 사이의 간격과 Contact Patch 중심 위치를 추출하려고 함
        따라서 주파수 간격 정보를 갖는 Y축이 그다지 필요가 없는 정보이고, 해당 구간에서 저주파 신호가 있었냐 없었냐가 중요
        이를 위해 2차원 이미지 정보에서 Y축을 모두 합하여 0일경우에는 0, 0이 아닌 경우에는 1로 표현하여 Contact Patch가 존재하는 구간이 1로 표현되게 함
        
    2-4. Contact Patch 구분 (알고리즘 구현해야함)
        Y축 병합까지 마쳤을 때, Contact Patch가 명확하게 보인다면 하나의 Contact Patch 신호가 존재하는 부분은 하나의 덩어리로 보일 것
        하지만 그건 이상적인 환경이고 그렇지 않은 경우도 있음.
        
        해당 알고리즘이 주파수 도메인 기반의 Contact Patch Extraction 알고리즘 이전에 2차 미분과 comparator sum을 적용한 알고리즘과 유사하게 적용할 수 있을 것으로 보여짐
        (Gitlab의 BANF_Collaboration_Work의 extractContactPatchSignal, windowSum, calculateComparator 참고)
        
    3. Contact Patch 구간에서의 고주파 신호 추출 (알고리즘 구현해야함)
        2의 알고리즘 구현에 의해 Contact Patch 위치의 시간 index를 알 수 있으면, 해당 정보를 기준으로 각 Contact Patch를 기준으로 이미지를 분할할 수 있음
        (현재 의견은 초기 3개의 Contact Patch 신호가 존재할 때, 가장 처음 Contact Patch 중간 위치 시간부터 가장 마지막 Contact Patch 중간 위치까지 자른 이미지를 사용하는 형태는 어떨 까 생각이 듬)
        (만약 초기 2개 Contact Patch일 경우와 마지막 2개 Contact Patch인 부분은 볼 수 있는 앞, 뒤 Contact Patch의 중앙부터 처음 혹은 마지막 구간까지를 보는 것으로)
        (Spectrogram 상 최소 2개의 Contact Patch가 존재해야 작동할 수 있음)
        (각 구간이 Overlap되게 하면 더 자세하게 볼 수 있을듯 함 (ex: Contact Patch 순서 기준 1, 2, 3 구간, 2, 3, 4 구간, 3, 4, 5 구간 ...))
        
    3-1. High Pass Filtering
        100Hz 이하 저주파는 Contact Patch 포함 시간을 찾아 구간을 분할하면서 그 역할을 다했기에, 여기에서는 100Hz 이상 High Pass Filtering을 적용
    
    3-2. 구간 단위로 구분된 이미지의 모든 픽셀의 합 계산
        High Pass Filter에 의해 100Hz 이상의 고주파 신호만 남은 이미지에서, 픽셀 값의 합을 구함
        이 합의 크기를 Threshold로 포트홀을 밟은 데이터가 포함된 구간과 포함되지 않은 구간을 구분할 수 있을 것으로 보임
        
    3-3 (?). 더 작은 구간으로 분할하기 위해 이진 탐색 적용
        3-2의 결과로 포트홀을 밟은 것으로 예상되는 시간에 해당하는 좁은 범위의 Spectrogram 이미지를 얻을 수 있음
        고주파 신호가 포함되는 더 좁은 구간으로 자르기 위해 전체 이미지를 시간에 대해 반으로 나눠, 두 분할 중 전체 픽셀 합이 더 큰 분할만 선택하면 포트홀을 밟은 시간대를 더 정확하게 파악할 수 있을 것으로 보임
        (하지만 이 가정은 해당 분할의 정 중앙에 포트홀을 밟았을 경우 모호해질 가능성이 있기에 좋은 방법은 아닐 수 있음)
        
        (3-3의 결과를 통해 왼쪽, 오른쪽 양쪽 간 차이 비교 시 3-3에 의해 특정된 해당 시간의 신호를 FFT로 주파수 분석하여 차이를 비교할 수 있을 것으로 예상됨)
        
        (알고리즘을 구성하면서 든 생각은 Spectrogram 상에서 고주파의 큰 신호가 존재하면 그냥 그 시간대에 포트홀을 밟은 것이니 그 알림만 주면 되지 않을까? 라는 생각도 들게됨)
        (이 생각대로면 3-3까지 한 작업과 특정된 그 시간대의 신호에 FFT를 적용한 결과 신호의 전체 면적을 기준으로 그 강도를 구분하여 알림을 줄 수 있을 것으로 예상됨)
    """

signal = [df["Acceleration_Z"] for df in data_list]

# Filtering 위한 변수 설정
fs = 1000
nyq = 0.5 * fs  # Calculate Nyquist frequency
order = 5
high = 100 / nyq

for sensor_idx, sig in enumerate(signal):
    plt.figure(figsize=(14, 5))
    # 2-1. Low Pass Filtering
    b, a = butter(order, high, btype='low')
    sig = filtfilt(b, a, sig)
    
    # Spectrogram 생성
    f, t, Sxx = spectrogram(sig, fs=1000, nperseg=nperseg, noverlap=noverlap)
    
    # Spectrogram 픽셀 색상 값을 0~255 사이 정수값으로 정규화
    Sxx_normalized = (
        (Sxx - Sxx.min()) / (Sxx.max() - Sxx.min()) * 255
    ).astype(np.uint8)

    # Colormap 적용
    colormap_img = cv2.applyColorMap(Sxx_normalized, cv2.COLORMAP_BONE)
    # 실제 Spectrogram은 y축 기준 아래가 0부터 시작하나 OpenCV의 이미지는 위가 0부터 시작하므로 세로로 뒤집음
    flipped_img = cv2.flip(colormap_img, 0)
    resized_img = cv2.resize(flipped_img, (len(sig), 500), interpolation=cv2.INTER_NEAREST)
    grayscale_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
    
    # Spectrogram 이미지 출력
    plt.imshow(grayscale_img, cmap='gray')
    plt.show()
    
    plt.figure(figsize=(14, 5))
    
    # 2-2. 픽셀 값 이진화
    _, binary_img = cv2.threshold(grayscale_img, 70, 255, cv2.THRESH_BINARY)
    
    plt.imshow(binary_img, cmap='gray')
    plt.show()
    
    # 2-3. Y축 병합
    binary_val = binary_img.sum(axis=0)
    binary_val = np.where(binary_val > 0, 1, 0)
    
    overlay = hv.Curve(None)
    overlay *= hv.Curve(binary_val)
    overlay.opts(
        opts.Curve(xlabel='Time(samples)', height=400, responsive=True, default_tools=['xpan,ypan,xbox_zoom,ybox_zoom,undo,redo,reset,save'], active_tools=['xbox_zoom'], toolbar='below', show_grid=True, show_legend=True),
    )
    show(hv.render(overlay))
```
